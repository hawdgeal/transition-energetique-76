```python
import pandas as pd
```

```python
df = pd.read_csv('dpe_data.csv')
```

```python
df['clean_dep'] = df['tv016_departement_code'].astype(str).str.zfill(2)
```

```python
df[['clean_dep','tv016_departement_code']].sort_values(by='clean_dep')
```

```python
df.nunique()
```

```python
cols = ["classe_consommation_energie",
    "tr001_modele_dpe_type_libelle",
    "annee_construction",
    "tr002_type_batiment_description",
    "date_etablissement_dpe",
    "clean_dep"]

df_dpe = df[cols]

df_dpe.info()
```

```python
df_dpe_bat = df_dpe.groupby(['classe_consommation_energie', 'clean_dep']).size().reset_index(name='nombre_batiments')
df_dpe_bat.info()
```

```python
df_dpe_bat.nunique()
```

```python
df_dpe_bat.sample(10)
```

```python
df_dpe_annee = df_dpe.sort_values(by=['clean_dep', 'classe_consommation_energie', 'annee_construction'])
df_dpe_annee.info()
```

```python
df_dpe_annee.head(100)
```

```python
df_dpe_bat.to_csv('grouped_batiments_par_classe_et_par_dep.csv', index=False)
df_dpe_annee.to_csv('grouped_batiments_par_classe_et_par_annee.csv', index=False)
```

```python
# Ensure the 'date_etablissement_dpe' column is a datetime type
df_dpe['date_etablissement_dpe'] = pd.to_datetime(df_dpe['date_etablissement_dpe'], errors='coerce')

# Extract the year from the 'date_etablissement_dpe' column
df_dpe['year_etablissement'] = df_dpe['date_etablissement_dpe'].dt.year

# Group by 'classe_consommation_energie', 'clean_dep', and 'year_etablissement'
grouped_df = (
    df_dpe.groupby(['classe_consommation_energie', 'clean_dep', 'year_etablissement'])
    .size()
    .reset_index(name='nombre_batiments')
)

# Optionally, sort the results
grouped_df = grouped_df.sort_values(by=['clean_dep', 'classe_consommation_energie', 'year_etablissement'])

# Export to CSV
grouped_df.to_csv('grouped_batiments_par_classe_et_par_annee_etablissement.csv', index=False)

# Confirm export
print("File exported as 'grouped_batiments_par_classe_et_par_annee_etablissement.csv'")
```
